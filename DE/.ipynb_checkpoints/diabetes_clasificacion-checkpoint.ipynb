```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# para las graficas
plt.style.use('seaborn-v0_8')
sns.set_palette("pastel")
```

Cargar datos

```python
df = pd.read_csv('C:\\Users\\manue\\Documents\\filiberto\\evaluacionExtracConBD\\datasets\\diabetes.csv')
print("Dataset diabetes cargado:", df.shape)
df.head()
```

```python
# 2. Revisamos si hay datos raros (ej: glucosa = 0?)
print(df.describe())

# Hay muchos 0 en columnas que no deberían tenerlos (Glucose, BloodPressure, etc.)
# Los reemplazamos esos 0 por la mediana
cols_con_ceros = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
for col in cols_con_ceros:
    df[col] = df[col].replace(0, df[col].median())

print("Valores 0 reemplazados por la mediana")
```

```python
# 2. Revisamos si hay datos raros (ej: glucosa = 0?)
print(df.describe())

# Hay muchos 0 en columnas que no deberían tenerlos (Glucose, BloodPressure, etc.)
# Los reemplazamos esos 0 por la mediana (es común en este dataset)
cols_con_ceros = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
for col in cols_con_ceros:
    df[col] = df[col].replace(0, df[col].median())

print("Valores 0 reemplazados por la mediana")
```

```python
# 3. Separamos X e y
X = df.drop('Outcome', axis=1)
y = df['Outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42, stratify=y)

print(f"Entrenamiento: {X_train.shape}, Prueba: {X_test.shape}")
```

```python
# 4. Justificación del algoritmo
print("""
JUSTIFICACIÓN:
Elegí Árbol de Decisión porque:
- Es muy fácil de interpretar (se puede ver el árbol)
- No necesita escalar las variables
- Maneja bien datos con valores faltantes o raros
- Es diferente a la Regresión Logística que usé en el supervisado
""")
```

```python
# 5. Búsqueda de hiperparámetros
param_grid = {
    'max_depth': [3, 4, 5, 6, 7, 8],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'criterion': ['gini', 'entropy']
}

tree = DecisionTreeClassifier(random_state=42)
grid = GridSearchCV(tree, param_grid, cv=5, scoring='accuracy')
grid.fit(X_train, y_train)

print("Mejores parámetros encontrados:")
print(grid.best_params_)
print(f"Mejor accuracy en validación: {grid.best_score_:.4f}")
```

```python
# 6. Modelo final y predicciones
best_tree = grid.best_estimator_
y_pred = best_tree.predict(X_test)

acc = accuracy_score(y_test, y_pred)
print(f"\nAccuracy en test: {acc:.4f}")
print("\nReporte completo:")
print(classification_report(y_test, y_pred, target_names=['No Diabetes', 'Diabetes']))
```

```python
# 7. Matriz de confusión bonita
plt.figure(figsize=(6,5))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=['No Diabetes', 'Diabetes'],
            yticklabels=['No Diabetes', 'Diabetes'])
plt.title('Matriz de Confusión - Árbol de Decisión')
plt.ylabel('Real')
plt.xlabel('Predicho')
plt.show()
```

```python
# 8. Gráfica del árbol
plt.figure(figsize=(20,12))
plot_tree(best_tree, 
          feature_names=X.columns,
          class_names=['No Diabetes', 'Diabetes'],
          filled=True, 
          rounded=True, 
          fontsize=12)
plt.title("Árbol de Decisión Optimizado - Diabetes", fontsize=18)
plt.show()
```
